```python
import pandas as pd
import numpy
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
traffic_20 = gpd.read_file("NYSDOT_TDV_AADT_2020.gdb.zip")
```

```python
traffic_20['Traffic_Station_Locations_COU'].unique()
```

```python
## New York County Code = 61
traffic_M20 = traffic_20[traffic_20.Traffic_Station_Locations_COU == 61.]
traffic_M20.head()
```

```python
traffic_M20.plot(legend=True,figsize=(30, 12))
plt.xlim(580000, 595000)
plt.ylim(4.501e6,4.53e6)
plt.axis('off')
plt.show()
```
